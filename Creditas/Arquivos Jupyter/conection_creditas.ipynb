### SEPARAR A PLANILHA EM TABELAS SQL 

A princípio vamos separar essa tabela, em 4 tabelas SQL, para simular a conexão com essas tabelas. 

```python
df1.head()
```

```python
conta = df1[['id', 'specific_type', 'created_at']]

usuario = df1[['id', 'gender', 'birth_date', 'phone_code', 'state']]

renda = df1[['id', 'cpf_restriction', 'monthly_income', 'registration_form_closed', 'declares_income_tax']]

automovel = df1[['id', 'brand', 'model_year', 'auto_value', 'auto_debt', 'loan_amount']]

operation_status = df1[['id', 'operation_status']]


```

Agora, vamos subir nossas tabelas para o nosso banco de dados. 

```python
engine_mysql = sqlalchemy.create_engine('mysql+pymysql://root:abc123@172.17.0.2:3306/Creditas')
```

```python
conta.to_sql("conta", engine_mysql, if_exists="replace",index=False)
```

```python
usuario.to_sql("usuario", engine_mysql, if_exists="replace",index=False)
```

```python
renda.to_sql("renda", engine_mysql, if_exists="replace",index=False)
```

```python
automovel.to_sql("automovel", engine_mysql, if_exists="replace",index=False)
```

```python
operation_status.to_sql("operation_status", engine_mysql, if_exists="replace",index=False)
```

```python
df2.to_sql("receita_empresa_estado", engine_mysql, if_exists="replace",index=False)
```
